```python
#Import punctuations and stopword list
import string
import nltk 
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
```

```python
#create empty lists and dictionaries
tweet = {}
number = []
fav = []
text = []
date = []
tweets = []

#copy is a solution I found online to solve my bug because I found out that everytime I append my tweet
#it overwrites the previous one

import copy
for index, element in enumerate(rawtweet):
    n=1
    #number
    if index % 4 == 0:
        tweet["number"] = element
    #fav
    if (index + 1) % 4 == 0:
        tweet["fav"] = element
        tweets.append(copy.deepcopy(tweet))
    #text
    if (index - 1) % 4 == 0:
    #make it lower
        elementlower = element.lower()
    #make it a string
        elementstring = elementlower.split()
    #get rid of punctuations
        cleanelementlist = []
        for word in elementstring:
    #get rid of stop words
            if word in stop_words:
                continue
    #get rid of handle names
            elif (word.startswith('@')):
                continue
     #get rid of punctuations
            else:
                for mark in string.punctuation:
                    word=word.replace(mark, '')
            cleanelementlist.append(word)
        #make it a string again because I want it to look neat in excel as one block
        cleantext = ' '.join(cleanelementlist)
        #clear http link
        linkstart = cleantext.find('http')
        #I found out that if there's no http linkstart will == -1
        if linkstart != -1:
            tweet["text"] = cleantext[:linkstart]
        else:
            tweet["text"] = cleantext

    #date
    if (index - 2) % 4 == 0:
        tweet["date"] = element
        
        
print(tweets)
```

```python
import nltk
nltk.download('vader_lexicon')
nltk.download('punkt')
```

```python
with open('job.txt') as f:
    lines = f.readlines()
```

```python
# first, we import the relevant modules from the NLTK library
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# next, we initialize VADER so we can use it within our Python script
analyzer = SentimentIntensityAnalyzer()

scoredict={'text':[], 'scores':[]}
# Then we call the polarity_scores method on our analyzer and pass it tweet1 
# This outputs a dictionary with negative, neutral, positive, and compound scores for the input text
for line in lines:
    words= line.split()
    if len(words) > 4:
        scores = analyzer.polarity_scores(line)
        scoredict['text'].append(line)
        scoredict['scores'].append(scores)
    
```

```python
print(scoredict)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(data=scoredict)
df['neg'] = df['text'].apply(lambda x:analyzer.polarity_scores(x)['neg'])
df['neu'] = df['text'].apply(lambda x:analyzer.polarity_scores(x)['neu'])
df['pos'] = df['text'].apply(lambda x:analyzer.polarity_scores(x)['pos'])
df['compound'] = df['text'].apply(lambda x:analyzer.polarity_scores(x)['compound'])
df.drop(['scores'], 1, inplace=True)
df.sort_values(by=['pos'], ascending=False)
```
